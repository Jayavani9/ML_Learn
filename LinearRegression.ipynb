Simple Linear Regression

```python
class LinearReg:
  def __init__(self):
    self.m = None
    self.b = None


  def fit(self, x, y):
    m = ((np.mean(x) * np.mean(y)) - np.mean(x * y)) / ((np.mean(x) ** 2) - np.mean(x ** 2))
    b = np.mean(y) - m * np.mean(x)
    self.m = m
    self.b = b

  def predict(self, x):
    return self.m * x + self.b


```

```python
import numpy as np

independentVar = np.array([1,2,3,4,5])
dependentVar = np.array([2,4,5,4,5])


linearReg = LinearReg()
linearReg.fit(independentVar, dependentVar)
print(linearReg.m)
print(linearReg.b)
print(linearReg.predict(independentVar))



```

```python
import matplotlib.pyplot as plt

plt.scatter(independentVar, dependentVar)
plt.plot(independentVar, linearReg.predict(independentVar))
plt.show()
```
